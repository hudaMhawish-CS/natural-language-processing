```python
import GetOldTweets3 as got
import nltk
import re
```

```python
# 1- Data Collecting
#جمع التغريدات من المملكة العربية السعودية
tweetCriteria = got.manager.TweetCriteria().setNear('saudi arabia')\
                                           .setSince("2019-11-19")\
                                           .setUntil("2019-11-21")\
                                           .setMaxTweets(500)
tweet = got.manager.TweetManager.getTweets(tweetCriteria)
# print(tweet)
for t in tweet:
    print(t.text)
```

```python
# تخزين التغريدات في ملف نصي
file = open("huda.txt", "w",encoding='utf-8')
for t in tweet:
    file.writelines(t.text)
#     out_file.writelines("\n")
file.close()
```

```python
file = open("huda.txt", "r",encoding='utf-8')
string=file.readlines()
file.close()
print(string)
```

```python
# 2- Text Cleaning and Normalization 
string_1=""
for i in string:
    string_1=string_1+i
pattern=r'(http\S+)|((\w(\d)?)+(.(\w(\d)?)+)?.com\S+)|(www\.[\s]+)'
t=re.sub(pattern,' ',string_1)  # (space) استبدال الروابط  ب
pattern_d=r'(.)\1+'
t = re.sub(pattern_d, r'\1', t)  #ازالة تكرارات الأحرف
pattern=r'@|,|;|؛|،|ٍٍ/|\(|\)|{|}|\[|\]|\|'
t=re.sub(pattern,' ',t) #استبدال محارف محددة
print(t)
```

```python
# توحيد شكل النص
# symbol الاحرف التي نريد استبدالها
# sumb_1 الاحرف التي نستبدل بها
symbol=["ئ","ة","أ","إ","آ","ؤ"] 
sumb_1=["ء","ه","ا","ا","ا","ء"]
for i in range(6):
    t = t.replace(symbol[i], sumb_1[i])
print(t)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
# 3- Text Analysis
#  stopwords  ازالة 
t_1=word_tokenize(t)
stopwords = nltk.corpus.stopwords.words('arabic')
t_2= [w for w in t_1 if w not in stopwords]
print(t_2)
```

```python
#الكلمات الشائعة
fdist =nltk.FreqDist(t_2);
print(fdist.most_common(50))
print("\n\n............................")
# الكلمات الأقل شيوعا
print(fdist.hapaxes()[:50])
print("\n\n............................")
# Collocations
from nltk import bigrams
print(list(bigrams(t_1)))
print("\n\n............................")
# hhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhh
```

```python
# 4- Trending Topics
# إيجاد الهاشتاغات الأكثر شيوعا
pattern=r'#(\S+_\S+)+'
# ايجاد كل الهاشتاغات
has=re.findall(pattern,t)
print(has)
# الهاشتاغ الاكثر تكرارا
fdist_h =nltk.FreqDist(has);
print("\n\n",fdist_h.most_common(1)," -->الهاشتاغ الاكثر تكرارا")
```
